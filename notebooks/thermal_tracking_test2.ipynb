```python
from ultralytics import YOLO

model_path = "/content/drive/MyDrive/MegaSet_Train/v2_safe_train/weights/last.pt"
data_path = f"{dataset.location}/data.yaml"

model = YOLO(model_path)

print(f"--- RESUMING TRAINING FROM CHECKPOINT ---")
print(f"Loading Model: {model_path}")
print(f"Using Data: {data_path}")

results = model.train(
    data=data_path,
    epochs=50,
    imgsz=640,
    batch=16,
    seed=0,
    deterministic=True,

    project='/content/drive/MyDrive/MegaSet_Train',
    name='v2_safe_train',

    # continue from where it stay
    resume=True
)
```
